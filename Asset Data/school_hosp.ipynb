```python
import geopandas as gpd
import pandas as pd
```

```python

shapefile_path = r"D:\ERASMUS\France_Sem3\side projects\APAD_oxford\Final_Dataset_Folder\bangladesh-latest-free.shp\gis_osm_pois_free_1.shp"
data = gpd.read_file(shapefile_path)

print(data.head())
```

```python
data.plot()
```

```python
unique_values = data['fclass'].unique()
print("Unique values in 'fclass':", unique_values)
```

```python
education_point = data[data['fclass'].isin(['school', 'college', 'kindergarten', 'university'])]
print(education_point)
```

```python
education_point.to_file("education_point.shp")
education_point.plot()
```

```python
healthcare_point = data[data['fclass'].isin(['hospital', 'clinic', 'doctors', 'veterinary', 'dentist', 'optician','nursing_home'])]
print(healthcare_point)
```

```python
healthcare_point.to_file("healthcare_point.shp")
healthcare_point.plot()
```
